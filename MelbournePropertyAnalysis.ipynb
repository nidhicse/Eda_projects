```python
df=df.set_index('Unnamed: 0')
```

```python
df=df.dropna().copy()
df.shape
```

```python
df=df.reset_index(drop=True).copy()
df.head(50)
```

```python
df["Date"]=pd.to_datetime(df["Date"],dayfirst=True,utc=True)
```

```python
df=df.loc[~df.duplicated(subset=["Address","Postcode","Propertycount"])].reset_index(drop=True).copy()
df
```

```python
df.shape
```

```python
#Finding the number of properties in each suburb
axes=df["Suburb"].value_counts().plot(kind="line",figsize=(20,10))
axes.set_xlabel("Suburbs")
axes.set_ylabel("No. of properties")
plt.show()
```

```python
#comparing the prices of propertiezs accross different regions
axes=df.groupby("Regionname")["Price"].agg("mean").plot(figsize=(20,6))
axes.set_xlabel("Region")
axes.set_ylabel("Average Price of property")
```

```python
#Looking at the method of the property
axes=df["Method"].value_counts().plot(kind="bar",title="Methods")
axes.set_xlabel("Method")
axes.set_ylabel("Count")
plt.show()
```

```python
#yearbuilt v/s Price
df_yearwise=pd.DataFrame()
for name,group in df.groupby("YearBuilt"):
    if df_yearwise.empty:
        df_yearwise=group[["Price"]].rename(columns={"Price":name})
    else:
        df_yearwise=df_yearwise.join(group[["Price"]].rename(columns={"Price":name}))
df_yearwise
```
