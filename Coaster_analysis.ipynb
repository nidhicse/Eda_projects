```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
df=pd.read_csv("coaster_db.csv")
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df=df[['coaster_name','Location','Manufacturer', 'Model', 'Duration', 'Capacity', 'Designer', 'Max vertical angle',
       'Drop',  'Fast Lane available', 'Closing date', 'Opened', 
        'Theme','year_introduced', 'latitude', 'longitude', 'Type_Main','opening_date_clean','speed_mph', 'height_ft',
       'Inversions_clean', 'Gforce_clean'
       ]].copy()
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
#converting into datetime format
df["opening_date_clean"]=pd.to_datetime(df["opening_date_clean"])
```

```python
#dropping an irrelevent column
df=df.drop(["Closing date"],axis=1)
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.rename(columns={'coaster_name':'Coaster_name','year_introduced':'Year_introduced', 'latitude':'Latitude',
       'longitude':'Longitude','opening_date_clean':'Opening_date', 'speed_mph':'Speed',
       'height_ft':'Height', 'Inversions_clean':'Inversions', 'Gforce_clean':'Gforce'})
```

```python
df.shape
```

```python
#removing the duplicate rows from the dataframe
df=df.loc[~df.duplicated(subset=["Coaster_name","Opening_date"])].reset_index(drop=True).copy()
```

```python
df["Year_introduced"].value_counts()
```

```python
ax=df["Year_introduced"].value_counts().head(10).plot(kind="bar",title="Top Years coasters introduced")
ax.set_xlabel("Year introduced")
ax.set_ylabel("count")
```

```python
ax=df["Speed"].plot(kind="hist",bins=20,title="Coaster speed(mph)")
ax.set_xlabel("Speed(mph)")
ax.set_ylabel("count")
```

```python
ax=df["Speed"].plot(kind="kde",title="Coaster speed(mph)")
ax.set_xlabel("Speed(mph)")
ax.set_ylabel("count")
```

```python
ax=df["Height"].plot(kind="hist",bins=40,title="Height of coasters")
ax.set_xlabel("Height in ft")
ax.set_ylabel("count")
```

```python
df.plot(kind="scatter",x="Height",y="Speed",title="Coaster Speed vs Height")
plt.show()
```

```python
sns.scatterplot(x="Height",y="Speed",data=df,hue="Year_introduced")
```

```python
sns.pairplot(df,vars=["Year_introduced","Height","Speed","Inversions","Gforce"],hue="Type_Main")
plt.show()
```

```python
df_corr=df[["Year_introduced","Height","Speed","Inversions","Gforce"]].dropna().corr()
```

```python
sns.heatmap(df_corr,annot=True)
```

```python
#What are the locations with the fastest rollercoasters(min of 10 coasters)?
df["Location"].value_counts()
```

```python
ax=df.query('Location!="Other"').groupby("Location")["Speed"].agg(["mean","count"]).query('count>=10')\
.sort_values("mean")["mean"].plot(kind="barh",figsize=(12,5),title="Average speed of coasters by location")
ax.set_xlabel("Average coaster speed")
plt.show()
```

```python
#In which years were the coasters with the fast speed introduced(more than 20 count of one speed)?
axes=df.groupby("Year_introduced")["Speed"].agg(["count","mean"]).query('count>=20').dropna().sort_values("mean")["mean"]\
.plot(kind="barh",figsize=(12,5),title="Fastest coasters introduced over the years")
axes.set_xlabel("Average speed of the coasters")
plt.show()
```
